```python
plt.figure(figsize=(6,3))
sns.countplot(x='Platform', data=df, palette='viridis' , hue='Platform')
plt.title('Number of Games by Platform')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(df['User Rating'], kde=True, color='skyblue', bins=20)
plt.title('Distribution of Game Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='Platform', y='User Rating', data=df, palette='Set2' , hue = 'Platform')
sns.stripplot(x='Platform', y='User Rating', data=df, color='black', alpha=0.5)
plt.title('Ratings by Platform with Individual Data Points')
plt.xlabel('Platform')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
platform_counts = df['Platform'].value_counts()
plt.figure(figsize=(3,3))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140)
plt.gca().add_artist(plt.Circle((0, 0), 0.5, color='white'))  # Create a donut shape
plt.title('Distribution of Games by Platform')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.violinplot(x='Platform', y='User Rating', data=df, palette='pastel', hue = 'Platform')
plt.title('Distribution of Ratings by Platform')
plt.xlabel('Platform')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
g = sns.FacetGrid(df, col='Platform', col_wrap=3, height=4, sharex=False)
g.map(sns.histplot, 'User Rating', kde=True, color='green')
g.set_titles('{col_name}')
plt.show()
```

```python
top_10_games = df.groupby('Game Title')['User Rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_games.values, y=top_10_games.index, palette='Blues_r')
plt.title('Top 10 Games by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Game')
plt.show()
```

Conclusion : Analyzing this dataset provided insights into the distribution of video game ratings and popular platforms. The visualizations revealed that certain platforms have higher average ratings, which could guide market strategies.
